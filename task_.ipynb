```python


import matplotlib.pyplot as plt

def plot_data(df):
    # Select numerical columns
    numerical_cols = df.select_dtypes(include=['int', 'float']).columns
    num_cols = len(numerical_cols)
    
    # Determine the number of rows and columns needed for subplots
    num_rows = (num_cols // 3) + (num_cols % 3 > 0)
    num_cols_per_row = min(num_cols, 3)  # Limit to 3 columns per row
    
    plt.figure(figsize=(5 * num_cols_per_row, 5 * num_rows))
    
    for i, col in enumerate(numerical_cols):
        plt.subplot(num_rows, num_cols_per_row, i + 1)
        df[col].plot(kind='hist', title=col, ax=plt.gca())
        plt.xlabel(col)
        plt.ylabel('Frequency')
    
    plt.tight_layout()  # Adjust layout to prevent overlap
    plt.show()

# Example usage
plot_data(df_a)
```
