```python
import numpy as np
import pandas as pd


# Load the dataset
df = pd.read_csv('tvmarketing.csv')
df.head()
```

```python
## kaikki komentorivit suomeksi

import matplotlib.pyplot as plt

X = df.drop(columns='Sales') # ennustajat
y = df['Sales'] # ennustettava
plt.scatter(X, y) # piirretään pisteet
plt.xlabel('TV advertising (1000$)') # x-akselin otsikko
plt.ylabel('Sales revenue (M$)') # y-akselin otsikko
plt.show() # näytetään kuvaaja

```

Vastemuuttuja ongelma, regressio ongelma. 
Lineaarinen regressio käyttö


```python
from sklearn.linear_model import LinearRegression


model = LinearRegression().fit (X, y) # sovitetaan malli

model.intercept_ ## selota kunnola mitä tämä tarkoittaa: mallin vakio (b0) eli suora leikkaa y-akselin kohdassa 7.03 miljoonaa dollaria (tässä tapauksessa)
beta0 = model.intercept_ # mallin vakio
beta1 = model.coef_[0]
beta1

```

```python
from sklearn.metrics import mean_absolute_error

preds = model.predict(X) # ennustetaan myynti
mean_absolute_error(y, preds) # lasketaan keskimääräinen virhe
```

```python
## miten tehdään ristiinvalidointi
from sklearn.model_selection import cross_val_score

cross_val_score(model, X, y, cv=5, scoring='neg_mean_absolute_error').mean()

```
