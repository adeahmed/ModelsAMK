```python
from sklearn.preprocessing import StandardScaler


scaler = StandardScaler()
X  = pd.DataFrame(scaler.fit_transform(X), columns=X.columns) ## kaikki on nyt keskitetty ja skaalattu. 
X
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.2, random_state=42)## jaetaan data train ja test setteihin

```

```python
import sklearn.metrics as metrics
from sklearn.neighbors import KNeighborsClassifier
def KNN_model(k):
  
    knn = KNeighborsClassifier(n_neighbors=k)
    
    knn.fit(X_train, Y_train.values.ravel())
    accuracy = metrics.accuracy_score(Y_test, knn.predict(X_test))
    recall = metrics.recall_score(Y_test, knn.predict(X_test), pos_label="M")
    precision = metrics.precision_score(Y_test, knn.predict(X_test),pos_label="M")
    return f'K: {k} Accuracy: {accuracy}, Recall: {recall}, Precision: {precision}'
```

```python
print(Y_test) ## testataan toimiiko koodi ja nähdään mitä pitäisi saada ulos 
print(set(Y_test)) ## nähdään mitä arvoja on 
## set: tekee listasta setin, joka poistaa duplikaatit.
## print(set(Y_test)) näyttää että arvoja on 2, 1 ja 0.

```

```python
print(KNN_model(3))
print(KNN_model(5))
print(KNN_model(7))
print(KNN_model(9))
print(KNN_model(11))
print(KNN_model(13))
print(KNN_model(15))
```

```python
from sklearn.metrics import confusion_matrix, ConfusionMatrixDisplay
from sklearn.neighbors import KNeighborsClassifier
model = KNeighborsClassifier(n_neighbors=9)
model.fit(X, Y.values.ravel())
Y_pred = model.predict(X_test)
cm = confusion_matrix(Y_test, Y_pred)
cmd = ConfusionMatrixDisplay(cm, display_labels=['M','B'])
cmd.plot()
```
